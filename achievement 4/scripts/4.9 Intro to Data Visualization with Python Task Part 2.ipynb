## Part 2

### Step 2

```python
df.head()
```

```python
# creating plot chart
bar=df['orders_day_of_week'].value_counts().plot.bar()
```

```python
hist= df['prices'].plot.hist(bins = 70)
```

```python
scatter=sns.scatterplot(x = 'prices', y = 'prices',data = df)
```

```python
# create a list holding True/False values to the test np.random.rant () <= 0.7
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big
big = df[dev]
```

```python
# Store 30% of the sample in the dataframe small
small= df[~dev]
```

```python
len(df)
```

```python
len(big) + len(small)
```

```python
# reducing sample to only neccessary columns
df_2 = small[['orders_day_of_week','prices']]
```

```python
# creating line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

### Step 3

```python
# creating histogram for order hour of day column
hist_order_hour_of_day= df['order_hour_of_day'].plot.hist(bins = 40)
```

#### Most orders are placed during the day around 8am-4pm reaching its peak hour at 10am. Slowest hours are between midnight and 6am.

### Step 4

```python
# creating a bar chart for loyalty_flag column
bar_loyalty_flag=df['loyalty_flag'].value_counts().plot.bar()
```

#### Most customers are 'regular customers', then second place is 'loyal customers'.

### Step 5

```python
# reducing sample to only neccessary columns
df_3 = small[['order_hour_of_day','prices']]
```

```python
# creating line chart
line_order_hod_prices = sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices')
```

#### Prices seem to be the highest at around 3am then drop significantly at 10 am. Prices rise back up again around 7pm.

### Step 6

```python
df_4 = small[['no_of_dependants','age']]
```

```python
#creating line chart to explore the connections between age and number of dependents
line_no_dependants_age = sns.lineplot(data = df_4, x = 'age',y = 'no_of_dependants')
```

#### This line chart shows no clear connection between age and number of dependents.

### Step 7

```python
#creating scatterplot to explore the connections between age and income
scatter_age_income=sns.scatterplot(x = 'age', y = 'income',data = df)
```

#### Ages 20-40 have a max income of about 400,000 where as ages 40+ have a max income of 600,000. Ages 40+ have more spending power.
