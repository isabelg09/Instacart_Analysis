```python
#subset first 1000000 rows
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
#user defined funcations
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#applying function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

## Step 1

```python
#using loc function
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range'] = 'Mid-range product'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range'] = 'Low-range product'
```

```python
df_ords_prods_merged['price_range'].value_counts(dropna=False)
```

```python
#using for-loops
df_ords_prods_merged['orders_day_of_week' ].value_counts (dropna = False)
```

```python
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged['Busiest_day'] = result
```

```python
df_ords_prods_merged['Busiest_day' ].value_counts (dropna = False)
```

```python
df_ords_prods_merged.head()
```

```python
#deleting extra columns
df_ords_prods_merged = df_ords_prods_merged.drop(columns = ['price_range_loc'])
```

## Step 2

```python
#creating new "busiest days" column
result2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value in [0,1]:
    result2.append("Busiest days")
  elif value in [3,4]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_days'] = result2
```

## Step 3

```python
#Checking values
df_ords_prods_merged['busiest_days' ].value_counts (dropna = False)
```

The busiest days and slowest days values increased significantly since we added an extra day to both. The values also add up correclty.

## Step 4

```python
#checking most popular hours
df_ords_prods_merged['order_hour_of_day' ].value_counts (dropna = False)
```

```python
#defining result3
result3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value>=9 and value<= 16:
    result3.append("Most orders")
  elif value<=6:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```

```python
df_ords_prods_merged['busiest_period_of_day'] = result3
```
