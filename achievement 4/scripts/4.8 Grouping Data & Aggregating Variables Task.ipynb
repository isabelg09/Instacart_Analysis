### Step 4

```python
# Creating a 'max_order' column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
# Creating loyalty flag column based on max order criteria
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking values
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

### Step 5

```python
# Check basis statistics of product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Step 6

```python
# Creating a 'mean_order' column
ords_prods_merge['mean_price_for_orders'] = ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
# Creating avg_spent column based on mean_order criteria
ords_prods_merge.loc[ords_prods_merge['mean_price_for_orders'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price_for_orders'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge.head(100)
```

### Step 7

```python
# Creating a median of 'days_since_prior_order' column
ords_prods_merge['median_of_days_since_prior_order'] = ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating avg_spent column based on mean_order criteria
ords_prods_merge.loc[ords_prods_merge['median_of_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_of_days_since_prior_order'] > 10) & (ords_prods_merge['median_of_days_since_prior_order'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_of_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# check frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
